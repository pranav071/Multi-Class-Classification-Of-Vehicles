# Image Classifier - <i><font color=orange>07 Dense networks</font></i>
We will build a 1-layer neural network with just the input and output layers. We will then build a 2-layer neural network with an additional hidden layer and compare the performance of the two.

```python
import os
# disable recognition of CUDA capable graphics card to avoid library updates. We will stick to CPU processing for this exercise.
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

import numpy as np
import pickle

import tensorflow as tf
```

```python
# load the processed lists and supporting data in npz format - we had earlier saved this in the first segment of our code.
with np.load('training.npz', allow_pickle=False) as npz_file:
    # Load the arrays
    imgs_train = npz_file['images']
    x_train = npz_file['features']
    y_train_ohe = npz_file['targets_ohe']
    y_train_int = npz_file['targets_int']
    labels_train = npz_file['labels']    
    
with np.load('validation.npz', allow_pickle=False) as npz_file:
    # Load the arrays
    imgs_valid = npz_file['images']
    x_valid = npz_file['features']
    y_valid_ohe = npz_file['targets_ohe']
    y_valid_int = npz_file['targets_int']
    labels_valid = npz_file['labels']    

with np.load('testing.npz', allow_pickle=False) as npz_file:
    # Load the arrays
    imgs_test = npz_file['images']
    x_test = npz_file['features']
    y_test_ohe = npz_file['targets_ohe']
    y_test_int = npz_file['targets_int']
    labels_test = npz_file['labels']    
    
with open('labels_dict.pkl', 'rb') as f:
    labels = pickle.load(f)
    
with open('results_dict.pkl', 'rb') as f:
    results = pickle.load(f)    
```

```python
# set a fixed seed to get the same results on every run for recording our findings for this exercise.
# ref: https://towardsdatascience.com/how-to-solve-randomness-in-an-artificial-neural-network-3befc4f27d45

from numpy.random import seed
from tensorflow import set_random_seed

seed(0)
set_random_seed(0)
```

<h3>Lets first build a 1-layer dense network with no hidden layer</h3>

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers
import tensorflow.keras as keras
from sklearn.preprocessing import StandardScaler

# Rescale train, validation and test data
scaler = StandardScaler()
X_train_rescaled = scaler.fit_transform(x_train)
X_valid_rescaled = scaler.transform(x_valid)
X_test_rescaled = scaler.transform(x_test)

# create only a 1 layer model consisting of the output layer with 6 logits representing the 6 vehicle types
model = Sequential()
model.add(Dense(6, activation='softmax'))

# Define loss function, optimizer and metrics to track during training
# we use the Stochastic gradient descent optimizer with the sparse_categorical_crossentropy loss function to work with our integer output data
model.compile(optimizer='sgd', loss='sparse_categorical_crossentropy', metrics=['acc'])

# End training when accuracy stops improving for more than 6 epochs
early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss', patience=6)

# Fit model - we will use our validation_data instead of validation_split of the training data
history = model.fit(
    x=X_train_rescaled, y=y_train_int,
    batch_size=32, epochs=50,
    callbacks=[early_stopping],
    validation_data=(X_valid_rescaled, y_valid_int),
    verbose=False,
    shuffle=False # Turn shuffle off to get the same results on every run.
)
```

```python
# lets plot the validation loss and accuracy curves

%matplotlib inline
import matplotlib.pyplot as plt

# Create two plots: one for the loss value, one for the accuracy
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

# Plot accuracy values
ax1.plot(history.history['loss'], label='train loss')
ax1.plot(history.history['val_loss'], label='val loss')
ax1.set_title('Validation loss {:.3f} (mean last 3)'.format(
    np.mean(history.history['val_loss'][-3:]) # last three values
))
ax1.set_xlabel('epoch')
ax1.set_ylabel('loss value')
ax1.legend()

# Plot accuracy values
ax2.plot(history.history['acc'], label='train acc')
ax2.plot(history.history['val_acc'], label='val acc')
ax2.set_title('Validation accuracy {:.3f} (mean last 3)'.format(
    np.mean(history.history['val_acc'][-3:]) # last three values
))
ax2.set_xlabel('epoch')
ax2.set_ylabel('accuracy')
ax2.legend()
plt.show()
```

```python
# evaluate the test data to find the test accuracy of our 1-layer neural network
(test_loss_1layer, test_accuracy_1layer) = model.evaluate(X_test_rescaled, y_test_int, batch_size=32)

print('Test loss: {:.2f}'.format(test_loss_1layer))
print('Test accuracy: {:.2f}%'.format(100*test_accuracy_1layer))
```

<h3>Now we will build a 2-layer dense network with 1 hidden layer</h3>

```python
# create only a 2 layer model with a hidden layer of 32 units and the output layer with 6 logits representing the 6 vehicle types
model = Sequential()
model.add(Dense(32, activation='relu', input_dim=1280))
model.add(Dense(6, activation='softmax'))

# Define loss function, optimizer and metrics to track during training
# we use the Stochastic gradient descent optimizer with the sparse_categorical_crossentropy loss function to work with our integer output data
model.compile(optimizer='sgd', loss='sparse_categorical_crossentropy', metrics=['acc'])

# Fit model - we will use our validation_data instead of validation_split of the training data
# Use the early_stopping function created earlier to end training when accuracy stops improving for more than 6 epochs
history = model.fit(
    x=X_train_rescaled, y=y_train_int,
    batch_size=32, epochs=50,
    callbacks=[early_stopping],
    validation_data=(X_valid_rescaled, y_valid_int),
    verbose=False,
    shuffle=False # Turn shuffle off to get the same results on every run.
)
```

```python
# lets plot the validation loss and accuracy curves

# Create two plots: one for the loss value, one for the accuracy
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(16, 4))

# Plot accuracy values
ax1.plot(history.history['loss'], label='train loss')
ax1.plot(history.history['val_loss'], label='val loss')
ax1.set_title('Validation loss {:.3f} (mean last 3)'.format(
    np.mean(history.history['val_loss'][-3:]) # last three values
))
ax1.set_xlabel('epoch')
ax1.set_ylabel('loss value')
ax1.legend()

# Plot accuracy values
ax2.plot(history.history['acc'], label='train acc')
ax2.plot(history.history['val_acc'], label='val acc')
ax2.set_title('Validation accuracy {:.3f} (mean last 3)'.format(
    np.mean(history.history['val_acc'][-3:]) # last three values
))
ax2.set_xlabel('epoch')
ax2.set_ylabel('accuracy')
ax2.legend()
plt.show()
```

```python
# evaluate the test data to find the test accuracy of our 2-layer neural network
(test_loss_2layer, test_accuracy_2layer) = model.evaluate(X_test_rescaled, y_test_int, batch_size=32)

print('Test loss: {:.2f}'.format(test_loss_2layer))
print('Test accuracy: {:.2f}%'.format(100*test_accuracy_2layer))
```

**Observation**: We get a test accuracy of **90%** for our **1-layer neural** network and a test accuracy of **98%** for our **2-layer neural network**. We can conclude from these findings that our 2-layer neural network works better than the 1-layer one.

```python
# save the accuracy metrics for our neural networks
results['1-layer nn'] = round(test_accuracy_1layer, 2)
results['2-layer nn'] = round(test_accuracy_2layer, 2)

# save the results dictionary later retrieval
with open('results_dict.pkl', 'wb') as f:
    pickle.dump(results, f)
```
