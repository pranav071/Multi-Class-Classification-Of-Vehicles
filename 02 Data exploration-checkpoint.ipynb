<h3>Lets plot 4 images from each category</h3>

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig,axes=plt.subplots(nrows=4,ncols=6,figsize=(18,12))

# we will depict 4 images per vehicle type 
for row in range(0, 4):
    for col in range(0, 6):
        # get the vehicle type as per the column index (0 - 6)
        vehicle_type = labels[col]
        
        # plot the image and set the title from the train set
        axes[row,col].imshow(imgs_train[labels_train == vehicle_type][row])
        axes[row,col].set_title(labels_train[labels_train == vehicle_type][row])
        
        # disable axis
        axes[row,col].set_axis_off()
       
plt.show()
```

<h3>Lets make a scree plot</h3>
The scree plot will depict a bar plot of each component with its explained variance along with a stepped graph with cumulative variance.

```python
#ref: clustering-and-dimensionality-reduction-v3/units/dimensionality-reduction-with-pca-v2
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Create Standard scaler
scaler = StandardScaler()

# Rescale data
X_rescaled = scaler.fit_transform(x_train)

# Create PCA transformer with all components
pca = PCA(n_components=None)

# Apply PCA to rescaled data
pca.fit(X_rescaled)

# Retrieve the proportion of variance explained (pve) value
pve = pca.explained_variance_ratio_

# create a large plot (double click to expand)
plt.figure(figsize=(84, 10))
plt.grid(True)

# Create bar plot3
xcor = np.arange(1, len(pve) + 1) # 1,2,..,n_components
plt.bar(xcor, pve)
plt.xticks(xcor)

# Add cumulative sum
pve_cumsum = np.cumsum(pve)
plt.step(
    xcor+0.5, # 1.5,2.5,..,n_components+0.5
    pve_cumsum, # Cumulative sum
    label='cumulative'
)

# Add labels
plt.title('Scree plot')
plt.xlabel('principal component')
plt.ylabel('proportion of variance explained')
plt.legend()
plt.show()
```

<h3>Lets identify how many PCA components explain 10%, 20%, , 90% and 100% of the variance</h3>
Though we can do this by visually observing the above scree plot or looking at the data from pve_cumsum,
we will get an accurate reading by writing a small loop using the data in pve_cumsum 

```python
# we will write a short loop to find out how many PCA components explain 10% to 100% variance
# we will use the pve_cumsum array which already has the cumulative sum of explained variance 

# variance_perc will be used to compare the cumulative sum of variance at every iteration to get values for 10%, 20% etc.
variance_perc = list(range(10, 101, 10))
# we will increment this index to get the next desired variance percentage once we identify a % match 
index = 0
# counter increments at every iteration to hold information of the number of components which explain the variance
counter = 1

for v in pve_cumsum:
    # multiply v by 100 and round to get percentage - pve_cumsum holds data between 0.0 and 1.0
    cumsum = round(v*100,0)
    if(cumsum >= variance_perc[index]):
        # if we find a match, print the number of components which explain the % variance
        print("%d%% (%.2f%%) of variance is explained by %d components." % (variance_perc[index], v*100, counter))
        # increment the index to start looking for the next % variance match
        index = index + 1
    # increment the counter to identify number of components for the next match
    counter = counter + 1
    # exit the loop if we have identified 100% of the variance
    if(cumsum >= 100):
        break
    
```

<h3>Lets visualize the features on a 2d-plot with PCA using the first two components.</h3>

```python
#ref: clustering-and-dimensionality-reduction-v3/units/dimensionality-reduction-with-pca-v2

# Apply PCA to rescaled data
X_train_pca  = pca.transform(X_rescaled)

# Plot each kind of object
for kind in list(labels.values()):
    idx = (labels_train == kind)

    # Plot their components
    plt.scatter(
        X_train_pca[idx, 0], X_train_pca[idx, 1],
        label='{}'.format(kind)
    )

# Labels and legend
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left") # display the legend outside the graph
plt.xlabel('1st component')
plt.ylabel('2nd component')
plt.show()
```

**Observation:**
- We already know that the first 2 components explain around 15% of the variance. 
- It is interesting to note that most of the different vehicle categories occupy different region of space.
- It is not surprising that category 'other' has more overlap as it may contain a variety of different vehicle types
