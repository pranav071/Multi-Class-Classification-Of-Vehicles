# Image Classifier - <i><font color=orange>04 Simple decision tree</font></i>
We will build a decision tree with a depth of 3 and visualize it using the graphviz library. We will then add PCA to the pipeline and tune its n value using GridSearch. We will compare the two approaches and record our findings.

```python
import os
# disable recognition of CUDA capable graphics card to avoid library updates. We will stick to CPU processing for this exercise.
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

import numpy as np
import pickle

import tensorflow as tf
```

```python
# load the processed lists and supporting data in npz format - we had earlier saved this in the first segment of our code.
with np.load('training.npz', allow_pickle=False) as npz_file:
    # Load the arrays
    imgs_train = npz_file['images']
    x_train = npz_file['features']
    y_train_ohe = npz_file['targets_ohe']
    y_train_int = npz_file['targets_int']
    labels_train = npz_file['labels']    
    
with np.load('validation.npz', allow_pickle=False) as npz_file:
    # Load the arrays
    imgs_valid = npz_file['images']
    x_valid = npz_file['features']
    y_valid_ohe = npz_file['targets_ohe']
    y_valid_int = npz_file['targets_int']
    labels_valid = npz_file['labels']    

with np.load('testing.npz', allow_pickle=False) as npz_file:
    # Load the arrays
    imgs_test = npz_file['images']
    x_test = npz_file['features']
    y_test_ohe = npz_file['targets_ohe']
    y_test_int = npz_file['targets_int']
    labels_test = npz_file['labels']    
    
with open('labels_dict.pkl', 'rb') as f:
    labels = pickle.load(f)
    
with open('results_dict.pkl', 'rb') as f:
    results = pickle.load(f)
```

<h3>Lets build a decision tree with a depth of 3</h3>
**NOTE:** We will use **integer encoded y** to avoid confusion between multiclass and multilabel classification.

```python
# ref: decision-trees-and-svms-v2/units/decision-trees-v2

from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA

# Create a decision tree pipeline
dt_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('dt', DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0))
])

# NOTE: Unlike knn, here we use the labels as y values instead. 
# This will help us in better visualizing the decision tree in graphviz 

dt_pipe.fit(x_train, y_train_int)

# Evaluate on train, validation and test set
train_accuracy = dt_pipe.score(x_train, y_train_int)
valid_accuracy = dt_pipe.score(x_valid, y_valid_int)
test_accuracy = dt_pipe.score(x_test, y_test_int)

print('Training Accuracy: {:.3f}'.format(train_accuracy))
print('Validation Accuracy: {:.3f}'.format(valid_accuracy)) 
print('Test Accuracy: {:.3f}'.format(test_accuracy)) 
```

<h3>Lets visualize the decision tree with graphviz</h3>

```python
from sklearn.tree import export_graphviz
import graphviz

# fetch the DecisionTreeClassifier object from the pipeline
dt = dt_pipe.named_steps['dt']

# Export decision tree
dot_data = export_graphviz(
    dt, out_file=None,
    feature_names=None, class_names=list(labels.values()),
    filled=True, rounded=True, proportion=False   
)

# Display decision tree
graphviz.Source(dot_data)
```

<h3>Lets see if we can improve the accuracy by using PCA</h3>

```python
import pandas as pd

# Create a decision tree pipeline with PCA
dt_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA()), 
    ('dt', DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0))
])

# Variable to store the results
gs_results = []

# Grid search upto 95 components (We saw in data exploration that 95 components explain roughly 80% of the variance)
for n in range(1, 96):
    dt_pipe.set_params(pca__n_components=n)
    dt_pipe.fit(x_train, y_train_int)

    # Save model and its performance on train/test sets
    gs_results.append({
        'n': n,
        'train_accuracy': dt_pipe.score(x_train, y_train_int),
        'test_accuracy': dt_pipe.score(x_valid, y_valid_int)
    })

# Convert results to DataFrame
gs_results = pd.DataFrame(gs_results)
gs_results.sort_values(by=['test_accuracy', 'n'], ascending=False).head()
```

```python
# get the best_n to tune the decision tree. We select best_n as the lowest n value with the the best test_accuracy

# start by getting the best test accuracy
best_n_test_accuracy = gs_results.sort_values(by='test_accuracy', ascending=False)['test_accuracy'].iloc[0]
# From the subset of our dataframe with the 'same best accuracy', pick the first N value 
# (which is incidentally also the lowest N value from the set of values with the best test accuracy).
# We do this, because we are likely to get multiple entries with the same best test accuracy and 
# picking just the top N value may be inefficient, so we go for the lowest N value.
best_n = gs_results[gs_results['test_accuracy'] == best_n_test_accuracy]['n'].iloc[0]

# set the best N as our n_components hyperparameter for PCA
dt_pipe.set_params(pca__n_components=best_n)

dt_pipe.fit(x_train, y_train_int)

# Evaluate on train, validation and test set
train_accuracy = dt_pipe.score(x_train, y_train_int)
valid_accuracy = dt_pipe.score(x_valid, y_valid_int)
test_accuracy = dt_pipe.score(x_test, y_test_int)

print('best_n: ', best_n)
print('Training Accuracy: {:.3f}'.format(train_accuracy))
print('Validation Accuracy: {:.3f}'.format(valid_accuracy)) 
print('Test Accuracy: {:.3f}'.format(test_accuracy)) 
```

**Conclusion:** We see that that our **test** accuracy significantly improves to **84%** when using **PCA=3** as compared to not using PCA where the test accuracy is **72%**.

```python
# save the accuracy metrics for decision trees
results['decision tree'] = test_accuracy

with open('results_dict.pkl', 'wb') as f:
    pickle.dump(results, f)
```
