# Image Classifier - <i><font color=orange>05 Logistic regression</font></i>
We will first evaluate a logistic regression without any hyperparameter tuning. We will then tune the regularization strength with cross validated grid search.
We will then pick 10 random test images, plot them and compute and list their probabilities for each class. 
Lastly, we will vary the regularization strength and re-compute probabilities on the same set of images to observe the effects.

```python
import os
# disable recognition of CUDA capable graphics card to avoid library updates. We will stick to CPU processing for this exercise.
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

import numpy as np
import pickle

import tensorflow as tf
```

```python
# load the processed lists and supporting data in npz format - we had earlier saved this in the first segment of our code.
with np.load('training.npz', allow_pickle=False) as npz_file:
    # Load the arrays
    imgs_train = npz_file['images']
    x_train = npz_file['features']
    y_train_ohe = npz_file['targets_ohe']
    y_train_int = npz_file['targets_int']
    labels_train = npz_file['labels']    
    
with np.load('validation.npz', allow_pickle=False) as npz_file:
    # Load the arrays
    imgs_valid = npz_file['images']
    x_valid = npz_file['features']
    y_valid_ohe = npz_file['targets_ohe']
    y_valid_int = npz_file['targets_int']
    labels_valid = npz_file['labels']    

with np.load('testing.npz', allow_pickle=False) as npz_file:
    # Load the arrays
    imgs_test = npz_file['images']
    x_test = npz_file['features']
    y_test_ohe = npz_file['targets_ohe']
    y_test_int = npz_file['targets_int']
    labels_test = npz_file['labels']    
    
with open('labels_dict.pkl', 'rb') as f:
    labels = pickle.load(f)
    
with open('results_dict.pkl', 'rb') as f:
    results = pickle.load(f) 
```

<h3> Lets evaluate a logistic regression model (without any hyperparameters tuning)</h3>

We will first create a logistic regression model using **softmax regression** with the 'saga' solver.

**NOTE:** LogisticRegression accepts integer encoded labels and we do not have a choice to provide one-hot-encoded ones.

```python
# ref: logistic-regressions-v2/units/scikit-learn-implementation-v2

from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# Create the pipeline. Here, we will use the softmax regression for multiclass classification.
logreg = Pipeline([
    ('scaler', StandardScaler()),
    ('logreg', LogisticRegression(multi_class='multinomial', solver='saga'))
])

logreg.fit(x_train, y_train_int)

# Evaluate on train, validation and test set
train_accuracy = logreg.score(x_train, y_train_int)
valid_accuracy = logreg.score(x_valid, y_valid_int)
test_accuracy = logreg.score(x_test, y_test_int)

print('Training Accuracy: {:.3f}'.format(train_accuracy))
print('Validation Accuracy: {:.3f}'.format(valid_accuracy)) 
print('Test Accuracy: {:.3f}'.format(test_accuracy)) 
```

Let us also create a logistic regression model using **One vs Rest strategy** with the 'liblinear' solver for comparison

```python

# Create the pipeline. Here, we will use the One vs Rest strategy for multiclass classification.
logreg = Pipeline([
    ('scaler', StandardScaler()),
    ('logreg', LogisticRegression(multi_class='ovr', solver='liblinear'))
])

logreg.fit(x_train, y_train_int)

# Evaluate on train, validation and test set
train_accuracy = logreg.score(x_train, y_train_int)
valid_accuracy = logreg.score(x_valid, y_valid_int)
test_accuracy = logreg.score(x_test, y_test_int)

print('Training Accuracy: {:.3f}'.format(train_accuracy))
print('Validation Accuracy: {:.3f}'.format(valid_accuracy)) 
print('Test Accuracy: {:.3f}'.format(test_accuracy)) 
```

**Observation:** The test accuracy is the same with both approaches on most runs. Sometimes, the OVR works slightly better. 

We will stick with 'ovr' pipeline while performing cross-validated grid-search in the following section.

<h3> Lets tune the regularization strength parameter with cross-validated grid-search</h3>
Since cross validation creates its own validation set, we could have merge our training and validation sets but we will not do so. This is to maintain parity with the training data used for the model without any hyperparamter tuning for a fair comparison.

```python
# merge validation data into training data for all further use in this segment.
#imgs_train = np.concatenate((imgs_train, imgs_valid), axis=0)
#x_train = np.concatenate((x_train, x_valid), axis=0)
#y_train_ohe = np.concatenate((y_train_ohe, y_valid_ohe), axis=0)
#y_train_int = np.concatenate((y_train_int, y_valid_int), axis=0)
#labels_train = np.concatenate((labels_train, labels_valid), axis=0)
```

```python
# ref: logistic-regressions-v2/units/grid-search-with-cross-validation-v2

from sklearn.model_selection import GridSearchCV

# Create a grid with 20 C values between 0.0001 and 1000 on a logarithmic scale as well as C=1
grid = {
    'logreg__C': np.append(np.logspace(-4, 4, num=20), 1.0)
}

# Create the GridSearchCV object with cv strategy set to to 5-fold cross-validation and the above grid.
# We use the same logreg object created earler with 'ovr' strategy and 'liblinear' solver.
grid_cv = GridSearchCV(logreg, grid, cv=5, return_train_score=True, verbose=True, n_jobs=8)

grid_cv.fit(x_train, y_train_int)

# Evaluate on train, validation and test set
train_accuracy_gs = grid_cv.score(x_train, y_train_int)
test_accuracy_gs = grid_cv.score(x_test, y_test_int)

print('Training Accuracy: {:.3f}'.format(train_accuracy_gs))
print('Test Accuracy: {:.3f}'.format(test_accuracy_gs)) 
```

```python
# let us use a dataframe to visualize the results
import pandas as pd

# Collect results in a DataFrame
cv_results = pd.DataFrame(grid_cv.cv_results_)

# Print a few interesting columns
cols = ['mean_test_score', 'std_test_score', 'mean_train_score', 'std_train_score', 'param_logreg__C']
cv_results[cols].sort_values('mean_test_score', ascending=False).head(10)
```

Lets plot the validation curves to see the performance of our grid search

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Plot training/validation accuracy curves
plt.semilogx(cv_results["param_logreg__C"], cv_results["mean_train_score"], label='training accuracy')
plt.semilogx(cv_results["param_logreg__C"], cv_results["mean_test_score"],  label='validation accuracy')
plt.xlabel('Regularization strength (C)')
plt.ylabel('Accuracy')
plt.title("Evolution of training/validation accuracy with $C$")
plt.legend()
plt.show()
```

**Observation:** 

Our cross validated grid search is able to find a C value that is better than the default (C=1) and thus provide better results all around. It gives us a test accuracy of **98%** as compared to **96%** of the model without any hyperparameter tuning.


<h3> Lets us now pick ten images and compute the probability for each class </h3>

```python
# lets pick 10 random image indices between 0 and 49 (since the test image set has 50 images)
random_test_image_indices = np.random.randint(0, 49, size=10)

# lets plot the 10 images
fig,axes=plt.subplots(nrows=1,ncols=10,figsize=(16,2))

for i in range(0, 10):
    # plot each neighbor and set the title to its distance to the randomly selected image
    axes[i].imshow(imgs_test[random_test_image_indices[i]])
    axes[i].set_title(labels_test[random_test_image_indices[i]])
    
    # disable axis
    axes[i].set_axis_off()

plt.show()

# print the C value of the best estimator
print("C: ", grid_cv.best_estimator_.named_steps['logreg'].C)

# call predict_proba by passing the 10 x_test values to retrieve an array of 10 entries
# note: GridCV already picks the best C value - which is then used by functions like predict_proba
p = grid_cv.predict_proba(x_test[random_test_image_indices])

# each array element is an array with 6 values for the 6 categories in our data set.
# we will also set the y_labels as the index 
pd.DataFrame(p, columns=list(labels.values()), index=labels_test[random_test_image_indices])

```

<h3> Lets vary the regularization strength and observe the effect on the probabilities </h3>

```python
# lets pick the C value with the lowest mean_test_score in expectation of a visibly worse performance
# for this we sort the dataframe by mean test score with the least  score on top and use the C value of the first entry.
C = cv_results[cols].sort_values('mean_test_score', ascending=True)['param_logreg__C'][0]
print("C: ", C)

# Create the pipeline. Here, we will use the One vs Rest strategy for multiclass classification.
logreg2 = Pipeline([
    ('scaler', StandardScaler()),
    ('logreg', LogisticRegression(C=C, multi_class='ovr', solver='liblinear'))
])

logreg2.fit(x_train, labels_train)

# call predict_proba by passing the 10 x_test values to retrieve an array of 10 entries
# note: GridCV already picks the best C value - which is then used by functions like predict_proba
p = logreg2.predict_proba(x_test[random_test_image_indices])

# each array element is an arrow with 6 values for the 6 categories in our data set.
# we will also set the y_labels as the index 
pd.DataFrame(p, columns=list(labels.values()), index=labels_test[random_test_image_indices])
```

**Observation**: By using stronger regularization (C=0.001) and running predict_proba on the same 10 values, we can see that the probabilities are much worse. Finding the best C value for our logistic regression is vital to a strong model.

```python
# save the accuracy metrics for logistic regression
# NOTE: We have chosen to use the test_accuracy of the model without hyperparameter tuning as it is slightly better than the one obtained from GridSearchCV
results['logistic'] = test_accuracy

with open('results_dict.pkl', 'wb') as f:
    pickle.dump(results, f)
```
